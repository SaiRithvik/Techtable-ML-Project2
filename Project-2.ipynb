```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
train_data1 = pd.read_csv('Train_AV3.csv')
```

```python
train_data1
```

```python
def encoding(df):
    df["Gender"].replace({"Male": 1, "Female": 0}, inplace=True)
    df['Married'].replace({'No': 0, 'Yes': 1}, inplace=True)
    df['Education'].replace({'Graduate': 1, 'Not Graduate': 0}, inplace = True)
    df['Self_Employed'].replace({'No': 0, 'Yes': 1}, inplace = True)
    df['Loan_Status'].replace({'Y': 1, 'N': 0}, inplace = True)
    df['Property_Area'].replace({'Urban': 0, 'Rural':1, 'Semiurban':2}, inplace= True)
encoding(train_data1)    
```

```python
train_data1
```

```python
def fillna_mean(df):
    for i in range(6,10):
        df[df.columns[i]] = df[df.columns[i]].fillna(df[df.columns[i]].sum()/df.shape[0])  ### Working
fillna_mean(train_data1)        
```

```python
train_data1
```

```python
def fillna_mode(df):    
    for i in range(1,6):
        df[df.columns[i]] = df[df.columns[i]].fillna(df[df.columns[i]].mode().values[0])
    df[df.columns[10]] = df[df.columns[10]].fillna(df[df.columns[10]].mode().values[0])
fillna_mode(train_data1)    
```

```python
train_data1.isna().any()
```

```python
train_data1
```

```python
def drop_Dependents(df):
    df1 = df.drop(['Dependents'], axis = 1)
    return(df1)
train_data1 = (drop_Dependents(train_data1))    
```

```python
X = train_data1.iloc[:,1:train_data1.shape[0]]
y = train_data1.iloc[:,train_data1.shape[1]-1:]
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.3) 
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(C=100,kernel='linear')
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
